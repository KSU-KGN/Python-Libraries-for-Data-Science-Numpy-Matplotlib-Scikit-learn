```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
# from sklearn.datasets import load_boston
# ImportError: `load_boston` has been removed from scikit-learn since version 1.2.
# т.к. "Boston House Prices dataset" не доступен, поработаем с аналогичной библиотекой
from sklearn.datasets import fetch_california_housing
```

```python
california = fetch_california_housing()
california.keys()
```

```python
print(california["DESCR"])
```

```python
data = california["data"]
feature_names = california["feature_names"]
X = pd.DataFrame(data, columns=feature_names)
X.head()
```

```python
target = california["target"]
Y = pd.DataFrame(target, columns=["price"])
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
y_pred_lr = lr.predict(X_test)
check_test_lr = pd.DataFrame({
    "Y_test": Y_test["price"],
    "Y_pred_lr": y_pred_lr.flatten()})
```

Для понимания оценим качество предсказаний модели lr (линейная регрессия):

```python
check_test_lr["Error"] = check_test_lr["Y_pred_lr"] - check_test_lr["Y_test"]
check_test_lr.head()
```

```python
from sklearn.metrics import mean_squared_error

mse_lr = mean_squared_error(check_test_lr["Y_pred_lr"], check_test_lr["Y_test"])
mse_lr
```

```python
from sklearn.metrics import r2_score

r2_lr = r2_score(check_test_lr["Y_test"],check_test_lr["Y_pred_lr"])
r2_lr
```
